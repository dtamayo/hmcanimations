```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(12,8))
    x = np.linspace(-1.5*np.pi, 1.5*np.pi, 100)
    #ax.plot(x, np.cos(x))
    #ax.plot(x, np.cos(2*x))
    #ax.plot(x, np.cos(3*x))
    ax.plot(x, -0.1*np.cos(x)+np.cos(2*x)+0.4*np.cos(3*x), lw=4)
    ax.set_xlabel('x', fontsize=32)
    ax.set_ylabel('V(x)', fontsize=32)
    ax.axes.get_xaxis().set_ticks([])
    ax.axes.get_yaxis().set_ticks([])
```

```python
fig, ax = plt.subplots(figsize=(12,8))
x = np.linspace(0, 2*np.pi, 100)
V = -0.1*np.cos(x)+np.cos(2*x) + 0.4*np.sin(x)
ax.plot(x, V, lw=4)
ax.set_xlabel(r'$\theta$', fontsize=32)
ax.set_ylabel(r'$V(\theta)$', fontsize=32)
ax.axes.get_xaxis().set_ticks([0, 2*np.pi])
ax.axes.get_yaxis().set_ticks([])
ax.set_xticklabels([0, r'$2\pi$'], fontsize=32)
ylim = ax.get_ylim()
```

```python
xmin = x[V == V.min()]
```

```python
fig, ax = plt.subplots(figsize=(12,8))
x = np.linspace(0, 2*np.pi, 100)
V = -0.1*np.cos(x)+np.cos(2*x) + 0.4*np.sin(x)
ax.plot(x, V, lw=4)
ax.set_xlabel(r'$\theta$', fontsize=32)
ax.set_ylabel(r'$V(\theta)$', fontsize=32)
ax.axes.get_xaxis().set_ticks([0, 2*np.pi])
ax.axes.get_yaxis().set_ticks([])
ax.set_xticklabels([0, r'$2\pi$'], fontsize=32)
ax.plot(x, 1.95*(x-xmin+0.02)**2 + V.min(), linestyle='--', lw=3)
ax.set_ylim(ylim)
ax.text(4.4, 1, r'$V \propto \theta^2$', fontsize=32, color='tab:orange')
```
